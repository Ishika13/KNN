```python
# getting datasets and reading them

dfx = pd.read_csv('https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/Datasets/xdata.csv')
dfx.head()
```

```python
# getting datasets and reading them

dfy = pd.read_csv('https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/Datasets/ydata.csv')
dfy.head()
```

```python
# choosing the values in the dataset

X = dfx.values
Y = dfy.values
X = X[:,1:]
Y = Y[:,1:].reshape((-1,))
```

```python
# printing datasets

print (X)
print (Y)
```

```python
# plotting the values in the dataset
# here we observe two distinct clusters 

plt.scatter(X[:,0],X[:,1])
```

```python
# finding a point and getting its class from the cluster
# in this case, the point belongs to the second cluster

query_x = np.array([2,3])
plt.scatter(query_x[0],query_x[1])
```

```python
# knn plots

def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    new_vals = np.unique(vals[:,1],return_counts=True)
  
    index = new_vals[1].argmax()
    point = new_vals[0][index]
    
    return point
```

```python
# getting values knn

x = knn(X,Y,[0,0])
print(x)
```
